```python
# Step 1: Install pandas profiling (now called ydata-profiling)
!pip install ydata-profiling seaborn pandas

# Step 2: Import libraries
import pandas as pd
import seaborn as sns
from ydata_profiling import ProfileReport

# Step 3: Load a built-in dataset (you can use 'titanic', 'iris', etc.)
df = sns.load_dataset("titanic")

# Step 4: Display the first few rows
df.head()

# Step 5: Create the profile report
profile = ProfileReport(df, title="Titanic Dataset Profiling Report", explorative=True)

# Step 6: Display the interactive report in Colab
profile.to_notebook_iframe()

# (Optional) Step 7: Export report to an HTML file
profile.to_file("titanic_profile_report.html")
```
