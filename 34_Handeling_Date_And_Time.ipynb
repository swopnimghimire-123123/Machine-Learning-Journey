### Working with Dates

```python
# Convert to datetime datatype
date["date"] = pd.to_datetime(date["date"])
```

```python
date.info()
```

```python
date.sample(2)
```

### 1. Extract Year

```python
date["date_year"] = date["date"].dt.year
date.sample(2)
```

### 2.Extract Month

```python
date["date_month_no"] = date["date"].dt.month
date.sample(2)
```

```python
date["date_month_name"] = date["date"].dt.month_name()
date.sample(2)
```

### Extract Days

```python
date["date_day"] = date["date"].dt.day
date.sample(2)
```

```python
# Day to week
date["Date_Day_of_week"] = date["date"].dt.dayofweek
date.sample(2)
```

```python
# Day of week - name
date["Date_Day_of_week_name"] = date["date"].dt.day_name()
date.drop(columns=["product_id", "city_id", "orders"]).sample(2)
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['Date_Day_of_week_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract weeks of the year

```python
date["Date_week"] = date["date"].dt.isocalendar().week
date.drop(columns = ["product_id","city_id","orders"]).head()
```

### Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract Semister

```python
date["semister"] = np.where(date["quarter"].isin([1,2]),1,2)
date.drop(columns=["product_id","city_id","orders"]).head()
```

### Extract Time elapsed between dates

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date["date"]
```

```python
( today - date["date"]).dt.days
```

```python
# Months passed

np.round((today -date['date']).dt.days / 30.44, 0)
```

```python
# Converting to datetime datatype
time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

```python
time["hour"] = time["date"].dt.hour
time["min"] = time["date"].dt.minute
time["sec"] = time["date"].dt.second

time.head()
```

### Extract Time part

```python
time["time"] = time["date"].dt.time

time.head()
```

### Time Difference

```python
today - time["date"]
```

```python
# In seconds
(today - time["date"])/np.timedelta64(1,"s")
```

```python
# In minute
(today - time["date"])/np.timedelta64(1,"m")
```

```python
# In hours
(today - time["date"])/np.timedelta64(1,"h")
```
