# Machine Learning Plan

Here is the plan to build and deploy a machine learning model:

1.  **Preprocessing + EDA + Feature Selection**: Clean and explore the data, identify relevant features.
2.  **Extract input and output cols**: Separate the features (input) from the target variable (output).
3.  **Scale the values**: Normalize or standardize the feature values.
4.  **Train test split**: Divide the data into training and testing sets.
5.  **Train the model**: Train a machine learning model on the training data.
6.  **Evaluate the model / model selection**: Assess the model's performance and select the best model.
7.  **Deploying the model**: Make the trained model available for predictions on new data.

```python
# preprocessing
df = df.iloc[:,1:]
```

```python
df.head()
```

```python
# EDA
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["cgpa"],df["iq"], c = df["placement"])
```

```python
# Extracting input and output cols
X = df.iloc[:,0:-1]
Y = df.iloc[:,-1]
```

```python
print(X)
```

```python
print(Y)
```

```python
# Train the split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train,X_test,Y_train,Y_test)
```

```python
# Scale the values
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = sc.transform(X_test)
```

```python
X_test
```

```python
# Training the model
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# Model training
clf.fit(X_train,Y_train)
```

```python
# Evaluate the model
Y_pred = clf.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
# plotting the decision boundary
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X = X_train, y = Y_train.values, clf = clf, legend=2)
```

```python
import pickle
```

```python
pickle.dump(clf,open("placement_model.pkl","wb"))
```
