# Working with CSV Files in Python

Comma Separated Values (CSV) is a simple file format used to store tabular data. You can easily work with CSV files in Python using the `pandas` library, which is a powerful tool for data manipulation and analysis.

## Reading a CSV File

To read a CSV file into a pandas DataFrame, you can use the `read_csv()` function:

### **1.Import pandas**

```python
import pandas as pd
```

### **Opening a local csv file**

```python
df = pd.read_csv("aug_train.csv")
df.head()
```

```python
df.shape
```

### **3. Opening a csv file from a URL**

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

### **4. Sep parameter**

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])
```

### **5. Index_col parameter**

```python
pd.read_csv("aug_train.csv",index_col='enrollee_id')
```

### **6. Header Parameter**

```python
pd.read_csv("test.csv")
```

```python
pd.read_csv("test.csv",header = 1)
```

###**7. use _cols parameter**

```python
pd.read_csv("aug_train.csv",usecols=["enrollee_id","gender","education_level"])
```

### **8. Squeese parameters**

```python
pd.read_csv("aug_train.csv",usecols=["gender"])["gender"]
```

## **9. Skiprows/nrows Parameter**

```python
# Learn more form the documentation
pd.read_csv("aug_train.csv",skiprows=[0,2])
```

```python
pd.read_csv("aug_train.csv",nrows=100)
```

### **9. Encoding parameter**

```python
pd.read_csv("zomato.csv",encoding="latin-1")
```

### **11. Skip bad lines**

```python
pd.read_csv("zomato.csv",encoding="latin-1",on_bad_lines='skip')
```

### **12. dtypes parameter**

```python
# converting the target from float to integer
pd.read_csv("aug_train.csv",dtype={"target":int})
```

### **13. Handeling Dates**

```python
pd.read_csv("/content/IPL Matches 2008-2020.csv",parse_dates=["date"])
```

### **14. Converters**

```python
pd.read_csv("/content/IPL Matches 2008-2020.csv")
```

```python
def rename(name):
  if name == "Royal Challengers Bangalore":
    return "RCB"
  else:
    return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv("/content/IPL Matches 2008-2020.csv",converters={"team1":rename})
```

### **15. na_value parameter**

```python
pd.read_csv("aug_train.csv",na_values=["Male",])
```

### **16. Loading a huge data in chunks**

```python
dfs = pd.read_csv("aug_train.csv",chunksize=5000)
```

```python
for chunks in dfs:
  print(chunks.shape)
```
