## Bivariate and Multivariate Analysis in EDA

Bivariate and multivariate analysis are crucial steps in Exploratory Data Analysis (EDA) for understanding relationships between variables.

### Bivariate Analysis

Bivariate analysis examines the relationship between two variables at a time. This helps to identify patterns, correlations, and potential dependencies between pairs of features in your dataset. Common techniques include:

*   **Scatter plots:** Visualizing the relationship between two numerical variables.
*   **Box plots or violin plots:** Comparing the distribution of a numerical variable across different categories of a categorical variable.
*   **Bar plots:** Comparing the counts or proportions of categories between two categorical variables.
*   **Correlation matrices:** Quantifying the linear relationship between pairs of numerical variables.

### Multivariate Analysis

Multivariate analysis explores the relationships among three or more variables simultaneously. This allows for a more comprehensive understanding of complex interactions and dependencies within the dataset. Techniques include:

*   **Pair plots:** Generating a grid of scatter plots for all pairs of numerical variables, often with histograms or kernel density estimates on the diagonal.
*   **3D scatter plots:** Visualizing the relationship between three numerical variables.
*   **Heatmaps:** Visualizing correlations between multiple numerical variables.
*   **Principal Component Analysis (PCA):** A dimensionality reduction technique that can reveal underlying patterns and relationships in high-dimensional data.
*   **Clustering:** Grouping data points based on their similarity across multiple variables.

By employing both bivariate and multivariate analysis techniques, you can gain valuable insights into your data, identify potential issues, and inform subsequent steps in your data analysis or modeling process.

## Load data

### Subtask:
Load a suitable built-in seaborn dataset for demonstration.


**Reasoning**:
Import necessary libraries and load the specified seaborn dataset.



```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')
display(df.head())
display(df.info())
```

## Scatter plot

### Subtask:
Create a scatter plot to visualize the relationship between two numerical variables.


**Reasoning**:
Create a scatter plot to visualize the relationship between 'total_bill' and 'tip' using seaborn.



```python
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='total_bill', y='tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.show()
```

## Bar plot

### Subtask:
Create a bar plot to visualize the relationship between a numerical and a categorical variable.


**Reasoning**:
Create a bar plot to visualize the relationship between the categorical variable 'day' and the numerical variable 'total_bill' using seaborn.



```python
sns.barplot(data=df, x='day', y='total_bill')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill ($)')
plt.title('Average Total Bill by Day of the Week')
plt.show()
```

## Box plot

### Subtask:
Create a box plot to visualize the distribution of a numerical variable across different categories of a categorical variable.


**Reasoning**:
Create a box plot to visualize the distribution of 'tip' across different 'day' categories, with appropriate labels and title.



```python
sns.boxplot(data=df, x='day', y='tip')
plt.xlabel('Day of the Week')
plt.ylabel('Tip ($)')
plt.title('Distribution of Tip Amounts by Day of the Week')
plt.show()
```

## Distribution plot

### Subtask:
Create a distribution plot to visualize the distribution of a numerical variable across different categories of a categorical variable.


**Reasoning**:
Create a distribution plot using `seaborn.histplot` to visualize the distribution of 'tip' by 'day', add appropriate labels and a title, and display the plot.



```python
sns.histplot(data=df, x='tip', hue='day', multiple='dodge', shrink=0.8)
plt.xlabel('Tip ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Tip Amounts by Day of the Week')
plt.show()
```

## Heatmap

### Subtask:
Create a heatmap to visualize the relationship between two categorical variables.


**Reasoning**:
Create a heatmap to visualize the relationship between the categorical variables 'day' and 'time' by first creating a contingency table and then plotting it as a heatmap with annotations and a title.



```python
contingency_table = pd.crosstab(df['day'], df['time'])
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Relationship between Day of the Week and Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Day of the Week')
plt.show()
```

## Clustermap

### Subtask:
Create a clustermap to visualize the relationship between two categorical variables and potential clustering.


**Reasoning**:
Generate a contingency table and create a clustermap from it to visualize the relationship between the two categorical variables and potential clustering. Add a title to the plot and display it.



```python
contingency_table = pd.crosstab(df['day'], df['time'])
clustermap = sns.clustermap(contingency_table, annot=True, fmt='d', cmap='Blues')
clustermap.fig.suptitle('Clustermap of Day of the Week and Time of Day', y=1.02)
plt.show()
```

## Pair plot

### Subtask:
Create a pair plot to visualize relationships between multiple numerical variables.


**Reasoning**:
Create a pair plot using the specified numerical columns from the DataFrame.



```python
sns.pairplot(df[['total_bill', 'tip', 'size']])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

## Line plot

### Subtask:
Create a line plot to visualize the relationship between two numerical variables, potentially over time or another ordered variable.


**Reasoning**:
Group the data by 'size', calculate the mean 'total_bill', and then create a line plot to visualize the relationship.



```python
avg_total_bill_by_size = df.groupby('size')['total_bill'].mean()
sns.lineplot(x=avg_total_bill_by_size.index, y=avg_total_bill_by_size.values)
plt.title('Average Total Bill by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Average Total Bill ($)')
plt.show()
```
