### Outlier Removal using Z-score

The Z-score is a measure of how many standard deviations away from the mean a particular data point is. It is calculated as:

$$ z = \frac{x - \mu}{\sigma} $$

Where:
- $x$ is the data point
- $\mu$ is the mean of the dataset
- $\sigma$ is the standard deviation of the dataset

A common practice is to consider any data point with a Z-score greater than 3 or less than -3 as an outlier. We can then remove these outliers from the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/42_placement (1).csv")
```

```python
df.sample(3)
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])

plt.show()
```

```python
print("Mean value of cgpa",df['cgpa'].mean())
print("Std value of cgpa",df['cgpa'].std())
print("Min value of cgpa",df['cgpa'].min())
print("Max value of cgpa",df['cgpa'].max())
```

```python
# Finding the boundary values
print("Highest allowed",df['cgpa'].mean() + 3*df['cgpa'].std())
print("Lowest allowed",df['cgpa'].mean() - 3*df['cgpa'].std())
```

```python
# Finding the outliers
df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.11)]
```

### Trimming

```python
# Trimming

new_df = df[(df['cgpa'] < 8.80) & (df['cgpa'] > 5.11)]
new_df
```

```python
# Approach 2

# Calculating the Zscore

df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean())/df['cgpa'].std()
```

```python
df.head()
```

```python
df[df['cgpa_zscore'] > 3]
```

```python
df[df['cgpa_zscore'] < -3]
```

```python
# Trimming
new_df = df[(df['cgpa_zscore'] < 3) & (df['cgpa_zscore'] > -3)]
```

```python
new_df
```

### Capping

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
lower_limit, upper_limit
```

```python
df["cgpa"] = np.where(
    df["cgpa"] > upper_limit,
    upper_limit,
    np.where(
        df["cgpa"] < lower_limit,
        lower_limit,
        df["cgpa"]
    )
 )
```
