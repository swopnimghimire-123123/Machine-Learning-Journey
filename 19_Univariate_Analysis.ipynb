## Univariate Analysis

Univariate analysis is the simplest form of analyzing data. It involves examining a single variable at a time. The primary goal of univariate analysis is to describe the data and find patterns within that single variable.

Common techniques used in univariate analysis include:

*   **Measures of Central Tendency:** Mean, median, and mode to describe the center of the data.
*   **Measures of Dispersion:** Range, variance, and standard deviation to describe the spread of the data.
*   **Frequency Distributions:** Tables or graphs (histograms, bar charts) to show how often each value or range of values appears.
*   **Box Plots:** To visualize the distribution of the data, including quartiles and potential outliers.

Univariate analysis is often the first step in data exploration before moving on to bivariate or multivariate analysis, which involve examining the relationships between two or more variables.

## Types of Data

Data can generally be classified into two main types:

### Numerical Data

Numerical data represents quantities that can be measured or counted. It can be further divided into:

*   **Discrete Data:** Represents counts and can only take specific, distinct values (e.g., number of students in a class, number of cars in a parking lot).
*   **Continuous Data:** Represents measurements and can take any value within a given range (e.g., height, weight, temperature).

### Categorical Data

Categorical data represents qualities or characteristics and can be divided into categories. It can be further divided into:

*   **Nominal Data:** Represents categories with no inherent order or ranking (e.g., colors, types of fruits, gender).
*   **Ordinal Data:** Represents categories with a natural order or ranking (e.g., education levels - high school, college, graduate; customer satisfaction ratings - poor, fair, good, excellent).

Understanding the type of data you are working with is crucial for choosing the appropriate statistical methods and visualizations for analysis.

**Reasoning**:
 Importing necessary libraries, loading the dataset, creating a DataFrame, and adding the target column, which can all be done in a single code block.



```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()

# Create a pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column
df['target'] = iris.target

# Display the first few rows of the DataFrame
display(df.head())
```

## Univariate analysis on numerical data

### Subtask:
Perform univariate analysis on the numerical columns (e.g., calculate mean, median, standard deviation, and visualize with histograms or box plots).


**Reasoning**:
Select the numerical columns, calculate descriptive statistics, and generate histograms and box plots for these columns.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numerical_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
df_numerical = df[numerical_cols]

# Calculate mean, median, and standard deviation
print("Descriptive Statistics for Numerical Columns:")
display(df_numerical.describe())

# Generate histograms
print("\nHistograms for Numerical Columns:")
df_numerical.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# Generate box plots
print("\nBox Plots for Numerical Columns:")
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_numerical)
plt.tight_layout()
plt.show()
```

## Univariate analysis on categorical data

### Subtask:
Perform univariate analysis on the categorical columns (e.g., calculate frequency counts and visualize with bar plots).


**Reasoning**:
Identify the categorical column, calculate its frequency counts, print the counts, and then generate and display a bar plot of these counts with appropriate labels and title.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify the categorical column
categorical_col = 'target'

# Calculate frequency counts
target_counts = df[categorical_col].value_counts()

# Print frequency counts
print("Frequency counts for the 'target' column:")
print(target_counts)

# Generate a bar plot of the frequency counts
plt.figure(figsize=(8, 6))
sns.barplot(x=target_counts.index, y=target_counts.values)

# Add title and axis labels
plt.title('Frequency Distribution of Iris Species (Target)')
plt.xlabel('Target (Species ID)')
plt.ylabel('Frequency')
plt.xticks(ticks=target_counts.index, labels=['setosa', 'versicolor', 'virginica']) # Assuming 0, 1, 2 correspond to species

# Display the bar plot
plt.show()
```
