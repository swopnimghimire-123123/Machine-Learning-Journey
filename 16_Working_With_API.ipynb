```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
# we have successfully extracted the things that we wanted from the first page
# now we need to do it for the next 428 pages
# it'll take some time to extract data from the 428 pages
for  i in range(1,429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
  temp_df = pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
  df = pd.concat([df, temp_df], ignore_index=True)
```
